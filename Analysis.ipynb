```python
import pandas as pd
data = pd.read_csv("C:/Users/joses/Downloads/myexcel.csv")
data
```

```python
data.isnull().sum()
```

```python
data['College'].fillna('new college')
```

```python
data['Salary'].fillna('10000')
```

```python
data.duplicated().sum()
```

```python
df = pd.DataFrame(data)

# Task 1: Distribution of employees across each team
team_distribution = df['Team'].value_counts()
total_employees = len(df)
team_percentage = (team_distribution / total_employees) * 100

print("Team Distribution:\n", team_distribution)
print("\nTeam Percentage:\n", team_percentage)
```

```python
# Task 2: Segregate employees based on their positions
position_distribution = df['Position'].value_counts()
print("\nPosition Distribution:\n", position_distribution)
```

```python
# Task 3: Identify the predominant age group among employees
age_groups = pd.cut(df['Age'], bins=[20, 25, 30, 35,40,45], labels=["20-25", "26-30", "31-35","36-40","41-45"])
age_group_distribution = age_groups.value_counts()
print("\nAge Group Distribution:\n", age_group_distribution)
```

```python
# Task 4: Highest salary expenditure by team and position
df['Salary'] = df['Salary'].fillna(0)
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
position_salary_expenditure = df.groupby('Position')['Salary'].sum()
highest_team_salary = team_salary_expenditure.idxmax()
highest_position_salary = position_salary_expenditure.idxmax()

print("\nTeam with Highest Salary Expenditure:", highest_team_salary)
print("Position with Highest Salary Expenditure:", highest_position_salary)
```

```python
# Task 5: Correlation between age and salary
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Relation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Task 1: Team Distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values, palette="viridis")
plt.title('Team Distribution')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
# Task 1: Team Distribution Percentage
plt.figure(figsize=(10, 6))
sns.barplot(x=team_percentage.index, y=team_percentage.values, palette="magma")
plt.title('Team Distribution Percentage')
plt.xlabel('Team')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

```python

# Task 2: Segregation of employees based on their positions
plt.figure(figsize=(5, 3))
sns.countplot(x='Position', data=df, palette="Set2")
plt.title('Position Distribution')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 3: Predominant age group among employees
plt.figure(figsize=(5, 3))
age_group_distribution.plot(kind='bar', color='red')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Task 4: Highest Salary Expenditure by Team and Position
plt.figure(figsize=(12, 6))
sns.barplot(x=team_salary_expenditure.index, y=team_salary_expenditure.values, palette="coolwarm")
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 4: Highest salary expenditure by team and position
plt.figure(figsize=(5, 3))
sns.barplot(x=position_salary_expenditure.index, y=position_salary_expenditure.values, palette="coolwarm")
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 5: Correlation between Age and Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, s=100, alpha=0.7, hue='Team')
plt.title('Relation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary ($)')
plt.legend(loc='best')
plt.show()
```
